```python
def is_safe(board, row, col)
    n = len(board) 
    for i in range(col):
        if board[row][i] == 1:
            return False
    for i, j in zip(range(row, -1, -1), range(col, -1, -1)):
        if board[i][j] == 1:
            return False
    for i, j in zip(range(row, n, 1), range(col, -1, -1)):
        if board[i][j] == 1:
            return False
    return True
def count_conflicts(board, row, col):
    n = len(board)
    conflicts = 0

   
    for i in range(n):
        if board[row][i] == 1 or board[i][col] == 1:
            conflicts += 1

   
    for i, j in zip(range(row, -1, -1), range(col, -1, -1)):
        if board[i][j] == 1:
            conflicts += 1
    for i, j in zip(range(row, n, 1), range(col, -1, -1)):
        if board[i][j] == 1:
            conflicts += 1

    return conflicts

def solve_queens(board, col):
    n = len(board)
    if col >= n:
        return True
    for i in range(n):
        if is_safe(board, i, col):
            board[i][col] = 1
            if solve_queens(board, col + 1):
                return True
            board[i][col] = 0

    return False

def eight_queens():
    n = 8  
    board = [[0] * n for _ in range(n)]

    if solve_queens(board, 0):
        print("Solution found:")
        for row in board:
            print(" ".join(["Q" if cell == 1 else "." for cell in row]))
    else:
        print("No solution exists.")

if __name__ == "__main__":
    eight_queens()
```

```python
import numpy as np
import itertools
class EightPuzzle:
    def __init__(self, square):
        self.state = square.copy()
        self.prev_state = None

    def printPath(self):
        if(self.prev_state != None):
            self.prev_state.printPath()
        print(self)

    def __repr__(self) -> str:
        string = ""
        for i, row in enumerate(self.state):
            string += f"{row[0]} - {row[1]} - {row[2]}\n"
            if i != 2:
                string += "|   |   |\n"
        return string
    
    def isSolved(self):
        return np.array_equal(self.state, np.array([[1, 2, 3], [8, 0, 4], [7, 6, 5]]))
    
    def estimatedCost(self, goal):
        cost = 0
        for i in range(3):
            for j in range(3):
                if(goal.state[i, j] == 0 or self.state[i, j] == 0):
                    pass
                elif(goal.state[i, j] == self.state[i, j]):
                    cost += 1
        return 8 - cost

    def findBlank(self):
        for i, row in enumerate(self.state):
            for j, ele in enumerate(row):
                if ele == 0:
                    return i, j
        print("thios is fucked up!!!")
        exit(-1)

    def swapStates(self, first, second):
        ele1 = self.state[first[0]][first[1]]
        ele2 = self.state[second[0]][second[1]]
        self.state[first[0]][first[1]] = ele2
        self.state[second[0]][second[1]] = ele1

    def genState(self, i):
        if(i == 1):
            return [i -1, i+1]
        elif(i == 0):
            return [i + 1]
        elif(i == 2):
            return [i-1]

    def nextStates(self):
        blank = self.findBlank()
        combinations = list(itertools.product([blank[0]], self.genState(blank[1]))) + list(itertools.product(self.genState(blank[0]), [blank[1]]))
        states = []
        for combination in combinations:
            sstate = EightPuzzle(self.state.copy())
            sstate.swapStates(blank, combination)
            sstate.prev_state = self
            states.append(sstate)
        return states

def solve(state : EightPuzzle, goal : EightPuzzle):
    visited = []
    stack = [state]
    i = 0
    def getcost(state : EightPuzzle):
        return state.estimatedCost(goal)
    while True:
        stack.sort(key = getcost)
        vertex = stack.pop(0)
        visited.append(vertex)
        if vertex.isSolved():
            print("The path is: ")
            vertex.printPath()
            break;
        for poss_state in vertex.nextStates():
            isin = False
            for statess in visited + stack:
                if np.array_equal(statess.state, poss_state.state):
                    isin = True
            if(not isin):
                stack.append(poss_state)
        i += 1

state = EightPuzzle(np.array([[2, 8, 3], [1, 6, 4], [7, 0, 5]]))
goal = EightPuzzle(np.array([[1, 2, 3], [4, 5, 6], [7, 0, 8]]))
solve(state, goal)
```

```python
from itertools import permutations
def solve_cryptarithmetic(puzzle:str):
    chars = set(puzzle.replace("+", "").replace("=", ""))
    for digits in permutations(range(10), len(chars)):
        mapping = dict(zip(chars, digits))
        equation = ""
        for char in puzzle:
            if char == "+" or char == "=":
                equation = equation + char
            else:
                equation = equation + str(mapping[char])
        first, second = equation.split("+")
        second, result = second.split("=")
        sol = int(first) + int(second) == int(result)
        if sol:
            print(mapping)
            exit(0)
            
puzzle = "CROSS+ROADS=DANGER"
solve_cryptarithmetic(puzzle)
```
